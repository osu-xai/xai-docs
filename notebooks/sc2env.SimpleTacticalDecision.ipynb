```python
import numpy as np
import sc2env
from sc2env.q_learning_agent import ConvNetQLearningAgent
from sc2env.environments.simple_towers import SimpleTowersEnvironment

# We create an environment, and a convolutional Q-learning agent
env = SimpleTowersEnvironment()
agent = ConvNetQLearningAgent(num_input_layers=18, num_actions=4)

# We initialize the environment to a default starting state
state = env.reset()
```

```python
import imutil
import PIL
# The state is a tuple of:
#   features_minimap: np array of features from the minimap view
#   features_screen: np array of features from the camera view
#   rgb_minimap: np array of an RGB pixel view of the minimap
#   rgb_screen: np array of RGB pixel rendered frame of Starcraft II
features_minimap, features_screen, rgb_minimap, rgb_screen = state
PIL.Image.fromarray(imutil.show(rgb_screen, return_pixels=True).astype(np.uint8))
```

```python
# Each episode consists of a single action, which results in the friendly units attacking one enemy group
# Our reward is the number of friendly units left at the end
action = agent.step(state)
state, reward, done, info = env.step(action)
loss = agent.update(reward)
if done:
    state = env.reset()
print('Agent took action {} and received reward {}'.format(action, reward))

features_minimap, features_screen, rgb_minimap, rgb_screen = state
PIL.Image.fromarray(imutil.show(rgb_screen, return_pixels=True).astype(np.uint8))
```
