Git Workflow
=================
for working with code
---------------------
For coding, we work on a branch, then when done push the branch and create a pull request on github
```
git clone https://github.com/osu-xai/someRepo.git   // pull the repo you want (if not already pulled)
git branch myFeatureBranch                          // create the branch you want to work on, if not 
                                                    // already created
git checkout myFeatureBranch                        // pull that branch into the "working tree"
```
- Make the changes you want
- then when you are ready to commit them...
```
git add *                                           // assuming you want to add everything that you 
                                                    // changed or created OR, git add each file 
                                                    // individually
git commit -m "some coherent, concise comment"                                                    
```
- run any automated tests on that branch to ensure you haven't broken anything:
```
git pull                                            // get the latest changes from the server
git merge master                                    // merge any changes from master into your working tree 
                                                    //for running tests WITH THE LATEST CODE FROM MASTER
appropriateTestCommand.sh
```
- if breaks, fix and then add, commit those changes.
- if doesn't break tests, then you can push
- if first time pushing your branch, then do...
```
git push -u origin myFeatureBranch                  // creates a copy of your local branch on the server 
                                                    // that can be used in a pull request. The -u ties 
                                                    // them together so that in future you can just
                                                    // say "git push"
```
- if this is not your first push of the branch, and you have already said "git push -u origin myFeatureBranch" then you can just do
```
git push
```
- then got to github and create a new pull request for merging your branch into master.  Since your branch includes everyone else's work that has already been put into master (due to your git pull and merge maneuver), the ONLY changes in play for the pull request are this most recent push of your code.


for creating notebooks  
----------------------
We use the master branch, and don't create our own branches of this repo so that we all have on one repo at meetings.
```
git clone https://github.com/osu-xai/xai-docs.git  // positions you on the master branch
cd xai-docs/workbooks
cp scenarioTemplate.ipynb myScenarioName.ipynb     // make a copy of the template with the desired name
jupyter workbook                                   // to launch the editor
```
In browser, navigate to XAI_SC2 and add a link to myScenarioName using markdown, then follow that link to edit the page. Once finished, commit, pull any changes that might have happened in the meantime and push your changes
```
git add *.ipynb
git commit -m "added myScenario workbook"
git pull
git push
```
